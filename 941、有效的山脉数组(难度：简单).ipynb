# 有效的山脉数组

**题目来源：力扣（LeetCode）**

**链接：https://leetcode-cn.com/problems/valid-mountain-array/**

## 一、题目

给定一个整数数组 A，如果它是有效的山脉数组就返回 true，否则返回 false。

让我们回顾一下，如果 A 满足下述条件，那么它是一个山脉数组：

- A.length >= 3
- 在 0 < i < A.length - 1 条件下，存在 i 使得：
    - A[0] < A[1] < ... A[i-1] < A[i]
    - A[i] > A[i+1] > ... > A[A.length - 1]
    


提示：

1. 0 <= A.length <= 10000
2. 0 <= A[i] <= 10000 

## 二、示例

### 示例 1：

输入：[2,1]

输出：false

### 示例 2：

输入：[3,5,5]

输出：false

### 示例 3：

输入：[0,3,2,1]

输出：true

## 三、解题思路

### 模拟过程

必须包含一个严格单调递增和严格单调递减的区间

```python
def validMountainArray(A) -> bool:
    n = len(A)
    if n < 3:
        return False
    top = -1
    for i in range(1, n):
        if top == -1:
            if A[i-1] > A[i]:
                if i - 1 < 1:
                    return False
                top = A[i-1]
            elif A[i-1] == A[i]:
                return False
        else:
            if A[i-1] <= A[i]:
                return False
    return top != -1
```

```python
validMountainArray([2,1])
```

```python
validMountainArray([3,5,5])
```

```python
validMountainArray([0,3,2,1])
```
