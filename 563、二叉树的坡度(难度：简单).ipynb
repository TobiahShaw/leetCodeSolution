```python
def findTilt(root) -> int:
    ret = 0
    def dfs(node):
        nonlocal ret
        if not node:
            return 0
        left = dfs(node.left)
        right = dfs(node.right)
        ret += abs(left - right)
        return node.val + left + right
    dfs(root)
    return ret
```
