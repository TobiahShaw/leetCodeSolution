# N叉树的最大深度

**题目来源：力扣（LeetCode）**

## 一、题目

给定一个 N 叉树，找到其最大深度。

最大深度是指从根节点到最远叶子节点的最长路径上的节点总数。

## 二、示例

例如，给定一个 3叉树 :



我们应返回其最大深度，3。

## 三、解题思路

```python
class Node:
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
```

```python
node_5 = Node(val=5, children=[])
node_6 = Node(val=6, children=[])

node_3 = Node(val=3, children=[node_5, node_6])
node_2 = Node(val=2, children=[])
node_4 = Node(val=4, children=[])

root = Node(val=1, children=[node_3, node_2, node_4])
```

### 递归

- 递归基为节点为None（返回0）或者节点的children为空（返回1）
- 否则递归的回去其所有child的maxDepth，选出最大的+1（节点本身一层）返回

```python
def maxDepth(root) -> int:
    if root is None: 
        return 0 
    elif root.children == []:
        return 1
    else: 
        height = [maxDepth(c) for c in root.children]
        return max(height) + 1
```
