```python
def maxDepthAfterSplit(seq):
    ret = []
    depth = 0
    for c in seq:
        if c == '(':
            depth += 1
            ret.append(depth % 2)
        elif c == ')':
            ret.append(depth % 2)
            depth -= 1
    return ret
```

```python
maxDepthAfterSplit(seq = "(()())")
```

```python
maxDepthAfterSplit(seq = "()(())()")
```
