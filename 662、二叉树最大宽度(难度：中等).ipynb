# 二叉树最大宽度

**题目来源：力扣（LeetCode）**

**链接：https://leetcode-cn.com/problems/maximum-width-of-binary-tree/**

## 一、题目

给定一个二叉树，编写一个函数来获取这个树的最大宽度。树的宽度是所有层中的最大宽度。这个二叉树与满二叉树（full binary tree）结构相同，但一些节点为空。

每一层的宽度被定义为两个端点（该层最左和最右的非空节点，两端点间的null节点也计入长度）之间的长度。

注意: 答案在32位有符号整数的表示范围内。

## 三、解题思路

### 一、层序遍历+trim（超时）

````python
def widthOfBinaryTree(root) -> int:
    def arr_trim(arr):
        if not arr:
            return arr
        left, right = 0, len(arr)
        while left < right and (arr[left] is None or arr[right - 1] is None):
            if not arr[left]:
                left += 1
            if not arr[right - 1]:
                right -= 1
        return arr[left:right]
    if not root:
        return 0
    ret = 0
    layer = [root]
    while layer:
        curr_layer = layer
        ret = max(len(curr_layer) if curr_layer else 0, ret)
        next_layer = []
        for node in curr_layer:
            if not node:
                next_layer.append(None)
                next_layer.append(None)
            else:
                next_layer.append(node.left)
                next_layer.append(node.right)
        next_layer = arr_trim(next_layer)
        layer = next_layer
    return ret
````

### 二、遍历

由于我们需要将给定树中的每个节点都访问一遍，我们需要遍历树。我们可以用深度优先搜索或者宽度优先搜索将树遍历。

这个问题中的主要想法是给每个节点一个 position 值，如果我们走向左子树，那么 position -> position * 2，如果我们走向右子树，那么 position -> positon * 2 + 1。当我们在看同一层深度的位置值 L 和 R 的时候，宽度就是 R - L + 1。

作者：LeetCode
链接：https://leetcode-cn.com/problems/maximum-width-of-binary-tree/solution/er-cha-shu-zui-da-kuan-du-by-leetcode/

````python
def widthOfBinaryTree(root) -> int:
    queue = [(root, 0, 0)]
    ret = 0
    cur_depth = 0
    layer_start = 0
    for node, depth, pos in queue:
        if not node:
            continue
        queue.append((node.left, depth + 1, pos * 2))
        queue.append((node.right, depth + 1, pos * 2 + 1))
        if cur_depth != depth:
            cur_depth =depth
            layer_start = pos
        ret = max(ret, pos - layer_start + 1)
    return ret
````
