### 

- 0 <= poured <= 10^9
- 0 <= query_glass <= query_row < 100

## 

###  1:

: poured() = 1, query_glass() = 1, query_row() = 1

: 0.00000

: 00

###  2:

: poured() = 2, query_glass() = 1, query_row() = 1

: 0.50000

: 001011

###  3:

: poured = 100000009, query_row = 33, query_glass = 17

: 1.00000

## 

### DP

```python
class Solution:
    def champagneTower(self, poured: int, query_row: int, query_glass: int) -> float:
        row =[poured]
        for i in range(1, query_row + 1):
            next_row = [0] * (i + 1)
            for j in range(len(row)):
                surplus = row[j] - 1
                if surplus > 0:
                    # 
                    next_row[j] += surplus / 2
                    next_row[j+1] += surplus / 2
            row = next_row
        return min(1, row[query_glass])
```

```python
s = Solution()
s.champagneTower(1, 1, 1)
```

```python
s.champagneTower(2, 1, 1)
```

```python
s.champagneTower(100000009, 33, 17)
```
