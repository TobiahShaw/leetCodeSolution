# 扁平化多级双向链表

**题目来源：力扣（LeetCode）**

**链接：https://leetcode-cn.com/problems/flatten-a-multilevel-doubly-linked-list/**

## 一、题目

多级双向链表中，除了指向下一个节点和前一个节点指针之外，它还有一个子链表指针，可能指向单独的双向链表。这些子列表也可能会有一个或多个自己的子项，依此类推，生成多级数据结构，如下面的示例所示。

给你位于列表第一级的头节点，请你扁平化列表，使所有结点出现在单级双链表中。

### 提示：

- 节点数目不超过 1000
- 1 <= Node.val <= 10^5

## 二、示例

### 示例 1：

输入：head = [1,2,3,4,5,6,null,null,null,7,8,9,10,null,null,11,12]

输出：[1,2,3,7,8,11,12,9,10,4,5,6]

解释：

```
1---2---3---4---5---6--NULL
         |
         7---8---9---10--NULL
             |
             11--12--NULL
```

序列化其中的每一级之后：

```
[1,2,3,4,5,6,null]
[7,8,9,10,null]
[11,12,null]
```

为了将每一级都序列化到一起，我们需要每一级中添加值为 null 的元素，以表示没有节点连接到上一级的上级节点。

```
[1,2,3,4,5,6,null]
[null,null,7,8,9,10,null]
[null,11,12,null]
```

合并所有序列化结果，并去除末尾的 null 。

[1,2,3,4,5,6,null,null,null,7,8,9,10,null,null,11,12]

### 示例 2：

### 示例 3：

## 三、解题思路

````python
class Node:
    def __init__(self, val, prev, next, child):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child
````

### DFS

````python
def flatten(head: 'Node') -> 'Node':
    if not head:
        return None
    def dfs(head, result):
        while head:
            result.append(head)
            if head.child:
                dfs(head.child, result)
            head = head.next
    ret = []
    dfs(head, ret)
    n = len(ret)
    for i in range(n):
        if i == 0:
            ret[i].prev = None
        else:
            ret[i].prev = ret[i-1]
        if i == n-1:
            ret[i].next = None
        else:
            ret[i].next = ret[i+1]
        ret[i].child = None
    return ret[0]
````
