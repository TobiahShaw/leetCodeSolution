# 将数据流变为多个不相交区间

**题目来源：力扣（LeetCode）**

**链接：https://leetcode-cn.com/problems/data-stream-as-disjoint-intervals/**

## 一、题目

给你一个由非负整数 a1, a2, ..., an 组成的数据流输入，请你将到目前为止看到的数字总结为不相交的区间列表。

实现 SummaryRanges 类：

- SummaryRanges() 使用一个空数据流初始化对象。
- void addNum(int val) 向数据流中加入整数 val 。
- int[][] getIntervals() 以不相交区间 [starti, endi] 的列表形式返回对数据流中整数的总结。

### 提示：

- 0 <= val <= 10^4
- 最多调用 addNum 和 getIntervals 方法 3 * 10^4 次
 

### 进阶：

如果存在大量合并，并且与数据流的大小相比，不相交区间的数量很小，该怎么办?

## 二、示例

输入：

["SummaryRanges", "addNum", "getIntervals", "addNum", "getIntervals", "addNum", "getIntervals", "addNum", "getIntervals", "addNum", "getIntervals"]

[[], [1], [], [3], [], [7], [], [2], [], [6], []]

输出：

[null, null, [[1, 1]], null, [[1, 1], [3, 3]], null, [[1, 1], [3, 3], [7, 7]], null, [[1, 3], [7, 7]], null, [[1, 3], [6, 7]]]

解释：

```
summaryRanges = SummaryRanges()
summaryRanges.addNum(1)      # arr = [1]
summaryRanges.getIntervals() # 返回 [[1, 1]]
summaryRanges.addNum(3)      # arr = [1, 3]
summaryRanges.getIntervals() # 返回 [[1, 1], [3, 3]]
summaryRanges.addNum(7)      # arr = [1, 3, 7]
summaryRanges.getIntervals() # 返回 [[1, 1], [3, 3], [7, 7]]
summaryRanges.addNum(2)      # arr = [1, 2, 3, 7]
summaryRanges.getIntervals() # 返回 [[1, 3], [7, 7]]
summaryRanges.addNum(6)      # arr = [1, 2, 3, 6, 7]
summaryRanges.getIntervals() # 返回 [[1, 3], [6, 7]]
```

## 三、解题思路

### 并查集

````python
class SummaryRanges:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self._set = set()

    def addNum(self, val: int) -> None:
        if val not in self._set:
            self._set.add(val)

    def getIntervals(self):
        l = list(self._set)
        l.sort()
        result = []
        if len(l) == 1:
            result.append([l[0], l[0]])
            return result
        left = right = 0
        for i in range(len(l) - 1):
            if l[i] + 1 == l[i + 1]:
                right = i + 1
                if i == len(l) - 2:
                    result.append([l[left], l[right]])
            else:
                if left == right:
                    if i != len(l) - 2:
                        result.append([l[left], l[left]])
                        left = right = i + 1
                    else:
                        result.append([l[left], l[left]])
                        result.append([l[left + 1], l[left + 1]])
                else:
                    if i != len(l) - 2:
                        result.append([l[left], l[right]])
                        left = right = i + 1
                    else:
                        result.append([l[left], l[right]])
                        result.append([l[right + 1], l[right + 1]])            
        return result
````

````python
summaryRanges = SummaryRanges()
print(summaryRanges.addNum(1))      # arr = [1]
print(summaryRanges.getIntervals()) # 返回 [[1, 1]]
print(summaryRanges.addNum(3))      # arr = [1, 3]
print(summaryRanges.getIntervals()) # 返回 [[1, 1], [3, 3]]
print(summaryRanges.addNum(7))      # arr = [1, 3, 7]
print(summaryRanges.getIntervals()) # 返回 [[1, 1], [3, 3], [7, 7]]
print(summaryRanges.addNum(2))      # arr = [1, 2, 3, 7]
print(summaryRanges.getIntervals()) # 返回 [[1, 3], [7, 7]]
print(summaryRanges.addNum(6))      # arr = [1, 2, 3, 6, 7]
print(summaryRanges.getIntervals()) # 返回 [[1, 3], [6, 7]]
````
